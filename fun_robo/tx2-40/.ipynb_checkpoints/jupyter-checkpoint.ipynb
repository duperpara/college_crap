```python
robtrix = Matrix([
        [Symbol('nx'), Symbol('ox'), Symbol('ax'), Symbol('px')],
        [Symbol('ny'), Symbol('oy'), Symbol('ay'), Symbol('py')],
        [Symbol('nz'), Symbol('oz'), Symbol('az'), Symbol('pz')],
        [0, 0, 0, 1],
    ])
display(robtrix)
```

```python
tot_transf_inv1 = (transformations[0].inv()*tot_transf).applyfunc(trigsimp)
tot_transf_inv2 = (transformations[1].inv()*tot_transf_inv1).applyfunc(trigsimp)
tot_transf_inv3 = (transformations[2].inv()*tot_transf_inv2).applyfunc(trigsimp)
tot_transf_inv4 = (transformations[3].inv()*tot_transf_inv3).applyfunc(trigsimp)
tot_transf_inv5 = (transformations[4].inv()*tot_transf_inv4).applyfunc(trigsimp)
tot_transf_inv6 = (transformations[5].inv()*tot_transf_inv5).applyfunc(trigsimp)

display(tot_transf_inv1)
display(tot_transf_inv2)
display(tot_transf_inv3)
display(tot_transf_inv4)
display(tot_transf_inv5)
display(tot_transf_inv6)
```

```python
robtrix_inv1 = (transformations[0].inv()*robtrix).applyfunc(trigsimp)
robtrix_inv2 = (transformations[1].inv()*robtrix_inv1).applyfunc(trigsimp)
robtrix_inv3 = (transformations[2].inv()*robtrix_inv2).applyfunc(trigsimp)
robtrix_inv4 = (transformations[3].inv()*robtrix_inv3).applyfunc(trigsimp)
robtrix_inv5 = (transformations[4].inv()*robtrix_inv4).applyfunc(trigsimp)
robtrix_inv6 = (transformations[5].inv()*robtrix_inv5).applyfunc(trigsimp)

display(robtrix_inv1)
display(robtrix_inv2)
display(robtrix_inv3)
display(robtrix_inv4)
display(robtrix_inv5)
display(robtrix_inv6)
```

```python
inv1_eq = (tot_transf_inv1 - robtrix_inv1).applyfunc(trigsimp)
inv2_eq = (tot_transf_inv2 - robtrix_inv2).applyfunc(trigsimp)
inv3_eq = (tot_transf_inv3 - robtrix_inv3).applyfunc(trigsimp)
inv4_eq = (tot_transf_inv4 - robtrix_inv4).applyfunc(trigsimp)
inv5_eq = (tot_transf_inv5 - robtrix_inv5).applyfunc(trigsimp)
inv6_eq = (tot_transf_inv6 - robtrix_inv6).applyfunc(trigsimp)

display(inv1_eq)
display(inv2_eq)
display(inv3_eq)
display(inv4_eq)
display(inv5_eq)
display(inv6_eq)
```

```python
all_eq_sys = [
    eq_system,
    inv1_eq,
    inv2_eq,
    inv3_eq,
    inv4_eq,
    inv5_eq,
    inv6_eq,
]
```

```python
theta_count_di = {}
for eq_sy_id, eq_sy in enumerate(all_eq_sys):
    for eq_id, eq in enumerate(eq_sy):
        theta_count_di[f"all_eq_sys[{eq_sy_id}][{eq_id}]"] = dict(sorted({str(symb): eq.count(symb) for symb in eq.free_symbols if 'θ' in str(symb)}.items()))
theta_count_di
```
